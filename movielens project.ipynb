```python
data1_=data['Genre'].str.get_dummies()
data2_=data['Gender'].str.get_dummies()
```

```python
data1_
```

```python
df1=pd.concat((data1_,data2_),axis=1)
df1

```

```python
df=data.iloc[:,0:7]
df

df2=pd.concat((df,df1),axis=1)
```

```python
df2=df2.drop(['Gender','Movie-Name'],axis=1)
```

```python
df2
```

```python
import seaborn as sns
sns.heatmap(df2.corr())
```

```python
df2.corr()
```

```python
X=df2.drop('Ratings',axis=1)
Y=df2['Ratings']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.5)
X_train.shape,Y_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
l1=LogisticRegression()
```

```python
l1.fit(X_train,Y_train)
```

```python
pred=l1.predict(X_test)
pred
```

```python
print("Train Accuracy",l1.score(X_train,Y_train)*100)
```

```python
print("Test Accuracy",l1.score(X_test,Y_test)*100)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_absolute_percentage_error
print('Mean Absolute Error:', mean_absolute_error(Y_test, pred).round(3))
print('Mean Squared Error:', mean_squared_error(Y_test, pred).round(3))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(Y_test, pred)).round(3))
print('Mean Absolute Percentage Error:',mean_absolute_percentage_error(Y_test,pred).round(3))
plt.figure(figsize=(7,5), dpi=75)
sns.regplot(Y_test, pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
d1=DecisionTreeClassifier(splitter='best')
```

```python
d1.fit(X_train,Y_train)
```

```python
pred1=d1.predict(X_test)
```

```python
pred1
```

```python
print("Train Accuracy",d1.score(X_train,Y_train)*100)
print("Test Accuracy",d1.score(X_test,Y_test)*100)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print('Mean Absolute Error:', mean_absolute_error(Y_test, pred1).round(3))
print('Mean Squared Error:', mean_squared_error(Y_test, pred1).round(3))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(Y_test, pred1)).round(3))
print('Mean Absolute Percentage Error:',mean_absolute_percentage_error(Y_test,pred1).round(3))
```
